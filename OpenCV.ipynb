# 2. Importing Libraries

```python
import cv2 as cv
import caer
import numpy as np
import matplotlib.pyplot as plt
```

# 3. Read Images

```python
# 1) Load image (color)

img = cv.imread('S:\Projects\Open CV\Open-CV\Images\895636 - Copy.jpg', cv.IMREAD_COLOR)

# 2) Load image (grayscale)
img_gray = cv.imread('S:\Projects\Open CV\Open-CV\Images\895636 - Copy.jpg', cv.IMREAD_GRAYSCALE)


# 3) Load image (unchanged, BGR)
plt.subplot(131)
plt.imshow(img)
plt.title('Original (BGR)')

# 4) Load image (grayscale)
plt.subplot(132)   
plt.imshow(img_gray, cmap='gray')
plt.title('Grayscale')


# 5) Load image (RGB)
plt.subplot(133)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('RGB')

plt.axis('off')
plt.tight_layout()
plt.show()
```

- An image is simply a 2D(grayscale) or 3D(color) Numpy Array

- Color images in OpenCV are stored in `BGR` order by default(not RGB)

```python
img = cv.imread('S:\Projects\Open CV\Open-CV\Images\895636 - Copy.jpg')

# Convert BGR to RGB
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# RGB image
plt.subplot(1,2,1)
plt.imshow(img_rgb)

#Default BGR image
plt.subplot(1,2,2)
plt.imshow(img)

plt.axis('off')
plt.show()
```

```python
import cv2

# 1) Load image (grayscale vs color)
img_color = cv2.imread("S:\Projects\Open CV\Open-CV\Images\895636 - Copy.jpg", cv2.IMREAD_COLOR)
img_gray  = cv2.imread("S:\Projects\Open CV\Open-CV\Images\895636 - Copy.jpg", cv2.IMREAD_GRAYSCALE)

# 2) Inspect properties
print("Color shape:", img_color.shape)     # e.g. (480, 640, 3)
print("Gray shape: ", img_gray.shape)      # e.g. (480, 640)
print("Dtype:      ", img_color.dtype)     # typically uint8

# 3) Display (in a notebook)

plt.subplot(121)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title("Color (RGB)")

plt.subplot(122)
plt.imshow(img_gray, cmap="gray")
plt.title("Grayscale")  

plt.axis("off")
plt.show()
```
